```python
import sympy as sp

def estimate_curve(grad_func, start_x, start_y, step, count):
    x_series = [start_x]
    y_series = [start_y]

    for _ in range(count):
        estimated_y = start_y + step * grad_func(start_x, start_y)
        start_x = round(start_x + step, 6)
        start_y = round(estimated_y, 6)
        x_series.append(start_x)
        y_series.append(start_y)

    return x_series, y_series


def launch_solver():
    print("Approximate a solution to dy/dx = f(x, y) using Euler's approach.\n")

    raw_equation = input("Define dy/dx in terms of x and y (e.g., x + y): ")
    x, y = sp.symbols('x y')
    expression = sp.sympify(raw_equation)
    slope_function = sp.lambdify((x, y), expression, modules='math')

    x_start = float(input("Start value of x: "))
    y_start = float(input("Start value of y: "))
    delta = float(input("Enter step length: "))
    rounds = int(input("How many steps to compute? "))

    x_path, y_path = estimate_curve(slope_function, x_start, y_start, delta, rounds)

    # Print to console
    print("\nIndex\tx\t\ty")
    for i in range(len(x_path)):
        print(f"{i}\t{x_path[i]}\t{y_path[i]}")

    # Save to output.txt
    with open("output.txt", "w") as file:
        file.write("Index\tx\ty\n")
        for i in range(len(x_path)):
            file.write(f"{i}\t{x_path[i]}\t{y_path[i]}\n")

    print("\nResults saved to output.txt ✅")


if __name__ == "__main__":
    launch_solver()
```
